```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ALL_dosage_UKB.csv")
```

```python
df.drop("Unnamed: 0", axis =1,inplace = True)
```

```python
weight_df = pd.read_excel("2020_05_05_CAD_PRS_267-converted.xlsx")
```

```python
weight_df.drop("Effect\nAllele", axis=1,inplace = True) 
```

```python
weight_df.drop("P-value", axis=1,inplace = True) 
weight_df.drop("Gene", axis=1,inplace = True) 
weight_df.drop("r2", axis=1,inplace = True) 
weight_df.drop("Genotyping", axis=1,inplace = True) 
weight_df.drop("LD", axis=1,inplace = True) 
weight_df.drop("rsID", axis=1,inplace = True) 
```

```python
df.drop("IID", axis = 1, inplace = True)





    
```

```python
def unweighted_score():
    
    total_columns = len(df.columns)

    unweighted_list = []
    for (rowName, rowData) in df.iterrows():
        total_score = rowData.values.sum()
        unweighted_score = total_score/ (total_columns * 2)
        unweighted_list.append(unweighted_score)
    return unweighted_list


def weighted_score():
    total_weight = weight_df["Weight"].values.sum()
    weight = weight_df["Weight"].values

    weightedscore_df = df.apply(lambda x: x * weight, axis = 1)
    weightedscore_df

    weighted_list = []
    for (rowName, rowData) in weightedscore_df.iterrows():
        total_score = rowData.values.sum()
        weighted_score = total_score/ (total_weight * 2)
        weighted_list.append(weighted_score)
    return weighted_list


unweighted_score()
weighted_score()
```

```python
new_df = pd.DataFrame(list(zip(unweighted_score(), weighted_score())), 
               columns =['Unweighted', 'Weighted']) 
```

```python
df.insert(0,"Unweighted", new_df["Unweighted"])
df.insert(1,"Weighted", new_df["Weighted"])
```
